### Answering Question 2

```python
tally = dataframe['Make'].value_counts()
tally
```

### Matplotlib Barplot

```python
# tally.plot(dataframe['Make'].value_counts())
tally.plot(kind='bar', title='Electric Vehicles Sold in Washington State', yticks=[0, 5000, 10000, 15000, 70000])
```

### Seaborn Barplot

```python
category = dataframe.Make.value_counts()  
category_df = category.reset_index()
category_df.columns = ['categories', 'frequency']
seaborn_plot = seaborn.barplot(x = 'categories', y = 'frequency', data = category_df)
for item in seaborn_plot.get_xticklabels():
    item.set_rotation(90)
```

```python
tally_model_year = dataframe['Model Year'].value_counts()
tally_model_year
```

### Matplotlib Barplot (line plot is not displaying properly, but it is the more appropriate type for this data)

```python
tally_model_year.plot(kind='bar',title='EVs Sold in each Model Year', yticks=[0, 5000, 10000, 15000, 25000, 40000])
```

### Seaborn Lineplot (more appropriate for the use case)

```python
category = dataframe['Model Year'].value_counts()  
category_df = category.reset_index()
category_df.columns = ['car sales', 'model year']
seaborn_plot = seaborn.lineplot(x = 'car sales', y = 'model year', data = category_df)
for item in seaborn_plot.get_xticklabels():
    item.set_rotation(90)
```
